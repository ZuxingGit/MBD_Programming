```python
# read data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('energydata_complete.csv')
df.head()
```

```python
# leave only the columns we need
df = df[['date', 'Appliances']]
df['date'] = pd.to_datetime(df['date'])
df = df.set_index('date')
df.head()
```

```python
# leave data at hh:10:00 and hh:40:00
df = df[df.index.minute.isin([10, 40])]
df.head(10)
```

```python
# take a sample data of one week
data_week = df['2016-02-01':'2016-02-07']
# take a sample data of 2 weeks
data_2weeks = df['2016-02-01':'2016-02-14']
# take a sample data of 4 weeks
data_4weeks = df['2016-02-01':'2016-02-28']
```
